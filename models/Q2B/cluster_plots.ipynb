```python
import pandas as pd
df = pd.read_csv('plotting.csv')
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting clusters
variables = ['Surface_Temp', 'Bottom_Temp', 'Surface_Salinity', 'Bottom_Salinity', 'temp_strat']
season_names = {
    'Aug_23': 'August 2023',
    # 'Jun_23': 'June 2023',
    'Feb_24': 'February 2024'
}
variable_names = {
    'Surface_Temp': 'Surface Temperature (Â°C)',
    'Bottom_Temp': 'Bottom Temperature (Â°C)',
    'Surface_Salinity': 'Surface Salinity (ppt)',
    'Bottom_Salinity': 'Bottom Salinity (ppt)',
    'temp_strat': 'Temperature Stratification'
}

sns.set(style="ticks")  # Minimal style with ticks but no grid

for season_code, season_name in season_names.items():
    season_df = df[df['sampling_bout'] == season_code]
    for variable, variable_name in variable_names.items():
        plt.figure(figsize=(10, 6))
        sns.boxplot(x='cluster', y=variable, data=season_df)
        plt.title(f'{variable_name} in {season_name}')
        plt.xlabel('Cluster')
        plt.ylabel(variable_name)
        plt.xticks(fontstyle='normal')  # Ensure cluster labels are not italicized
        sns.despine()  # Remove the top and right spines
        plt.tight_layout()
        plt.show()
```
