```python
import pandas as pd
df = pd.read_csv('plotting.csv')
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Plotting clusters
variables = ['Surface_Temp', 'Bottom_Temp', 'Surface_Salinity', 'Bottom_Salinity', 'temp_strat']
seasons = ['Aug_23', 'Jun_23', 'Feb_24']

for season in seasons:
    season_df = df[df['sampling_bout'] == season]
    for variable in variables:
        plt.figure(figsize=(10, 6))
        sns.boxplot(x='cluster', y=variable, data=season_df)
        plt.title(f'{variable} in {season}')
        plt.xlabel('Cluster')
        plt.ylabel(variable)
        plt.show()
```
