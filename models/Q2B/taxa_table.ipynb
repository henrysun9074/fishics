```python
import pandas as pd
df = pd.read_csv('plotting.csv')

feb_df = df[df['sampling_bout'] == 'Feb_24']
feb_df = feb_df.iloc[:, 12:]
feb_df

aug_df = df[df['sampling_bout'] == 'Aug_23']
aug_df = aug_df.iloc[:,12:]
aug_df
```

```python
non_species_columns = [
    'Longitude', 'Latitude', 'Surface_Temp', 'Bottom_Temp',
    'Surface_Salinity', 'Bottom_Salinity', 'temp_strat',
    'salt_strat', 'tSNE1', 'tSNE2', 'cluster'
]

# Identify species columns by excluding non-species columns
species_columns = feb_df.columns.difference(non_species_columns)
cluster_sums = feb_df.groupby('cluster')[species_columns].sum()
def get_top_taxa(cluster_sums, top_n=5):
    top_taxa = {}
    for cluster in cluster_sums.index:
        top_taxa[cluster] = cluster_sums.loc[cluster].nlargest(top_n)
    return top_taxa

# Get top taxa for each cluster
top_taxa_per_cluster = get_top_taxa(cluster_sums)

# Display the top taxa for each cluster
for cluster, taxa in top_taxa_per_cluster.items():
    print(f"Top taxa in cluster {cluster}:\n{taxa}\n")
top_taxa_df = pd.DataFrame.from_dict(top_taxa_per_cluster, orient='index')
top_taxa_df
```
